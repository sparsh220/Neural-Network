## Question 1: What is Deep Learning? Briefly describe how it evolved and how it differs from traditional machine  learning.

Answer: Deep learning is a part of machine learning. It uses neural networks with many layers to learn from data, like images, text, and sound.

How it evolved

* First came simple machine learning (rules + basic models).

* Then neural networks were introduced, but computers were slow.

* Later, big data + fast GPUs made deep learning powerful.

* Now it is used in voice assistants, self-driving cars, and translation.

Difference from traditional machine learning

* Traditional ML: humans give features (important data parts).

* Deep learning: model finds features by itself.

* Traditional ML works well on small data.

* Deep learning needs large data and more power.


## Question 2: Explain the basic architecture and functioning of a Perceptron. What are its limitations?


Answer: A Perceptron is the simplest neural network model. It is used for binary classification (yes/no type).

Basic architecture are :

* Inputs (x₁, x₂, …): data values

* Weights (w₁, w₂, …): importance of each input

* Bias (b): shifts the result

* Summation: multiply inputs with weights and add them

* Activation function: usually a step function

Limitations

* Can solve only linearly separable problems

* Cannot learn XOR problem

* Works only for binary output

* Very simple, not good for complex tasks


## Question 3: Describe the purpose of activation function in neural networks. Compare Sigmoid, ReLU, and Tanh functions.


Answer: An activation function decides how a neuron gives output.
Its main purpose is to add non-linearity, so the network can learn complex patterns.
Without it, a neural network becomes just a linear model.

* Sigmoid

    * Output range: 0 to 1

     * Used in binary classification

    * Problem: vanishing gradient (learning becomes slow)

* ReLU (Rectified Linear Unit)

    * Output range: 0 to infinity

     * Very fast and simple

    * Problem: dead neurons (output stays 0)

* Tanh

    * Output range: –1 to 1

    * Zero-centered, better than sigmoid

    * Still has vanishing gradient problem




## Question 4: What is the difference between Loss function and Cost function in neural networks? Provide examples.

Answer: Difference between Loss function and Cost function

*  Loss Function

    * It measures error for one training example.

    * Used to see how wrong the model is for one data point.

    * Example:

    *  Mean Squared Error (for one sample)

     * Binary Cross-Entropy (for one sample)

* Cost Function

    * It measures average error for all training examples.

    * It is the overall loss of the model.

    * The model tries to minimize the cost function during training.

    * Example:

     * Mean Squared Error (for all samples)

    * Cross-Entropy Cost


## Question 5: What is the role of optimizers in neural networks? Compare Gradient Descent, Adam, and RMSprop.


Answer:  Role of Optimizers in Neural Networks

* Optimizers change the weights of the network.

* They help reduce the loss.

* Goal: find best weights faster and correctly.

compare Gradient Descent, Adam, and RMSprop are :

* Gradient Descent (GD)

    * Updates weights using full data.

    * Slow for big datasets.

    * Simple and stable.

    * Example  : Move step by step to reach the lowest point.


* Adam (Adaptive Moment Estimation)

    * Uses learning rate + momentum.

    * Very fast and popular.

    * Works well for most problems.

* RMSprop

    * Changes learning rate automatically.

    * Good for non-stationary data.

    * Faster than Gradient Descent.

```python
# Question 6: Write a Python program to implement a single-layer perceptron from scratch using NumPy to solve the logical AND gate.
# (Include your Python code and output in the code box below.)

import numpy as np

# AND gate input and output
X = np.array([
    [0, 0],
    [0, 1],
    [1, 0],
    [1, 1]
])

y = np.array([0, 0, 0, 1])

# Initialize weights and bias
weights = np.zeros(2)
bias = 0
lr = 0.1

# Step function
def step_function(x):
    return 1 if x >= 0 else 0

# Training
for epoch in range(10):
    for i in range(len(X)):
        linear_output = np.dot(X[i], weights) + bias
        y_pred = step_function(linear_output)
        error = y[i] - y_pred

        weights = weights + lr * error * X[i]
        bias = bias + lr * error

# Testing
print("AND Gate Output:")
for i in range(len(X)):
    output = step_function(np.dot(X[i], weights) + bias)
    print(X[i], "->", output)

```

```python
# Question 7: Implement and visualize Sigmoid, ReLU, and Tanh activation functions using Matplotlib.
# (Include your Python code and output in the code box below.)

import numpy as np
import matplotlib.pyplot as plt

# Input values
x = np.linspace(-10, 10, 200)

# Activation functions
sigmoid = 1 / (1 + np.exp(-x))
relu = np.maximum(0, x)
tanh = np.tanh(x)

# Sigmoid
plt.figure()
plt.plot(x, sigmoid)
plt.title("Sigmoid Activation Function")
plt.xlabel("x")
plt.ylabel("Sigmoid(x)")
plt.show()

# ReLU
plt.figure()
plt.plot(x, relu)
plt.title("ReLU Activation Function")
plt.xlabel("x")
plt.ylabel("ReLU(x)")
plt.show()

# Tanh
plt.figure()
plt.plot(x, tanh)
plt.title("Tanh Activation Function")
plt.xlabel("x")
plt.ylabel("Tanh(x)")
plt.show()

```

```python
# Question 8: Use Keras to build and train a simple multilayer neural network on the MNIST digits dataset.
# Print the training accuracy.
# (Include your Python code and output in the code box below.)

import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten
from tensorflow.keras.datasets import mnist

# Load MNIST data
(x_train, y_train), (x_test, y_test) = mnist.load_data()

# Normalize values
x_train = x_train / 255.0
x_test = x_test / 255.0

# Build neural network
model = Sequential([
    Flatten(input_shape=(28, 28)),
    Dense(128, activation='relu'),
    Dense(10, activation='softmax')
])

# Compile model
model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

# Train model
history = model.fit(x_train, y_train, epochs=5, batch_size=32)

# Print training accuracy
print("Training Accuracy:", history.history['accuracy'][-1])
```

```python
# Question 9: Visualize the loss and accuracy curves for a neural network model trained on the Fashion MNIST dataset.
# Interpret the training behavior.
# (Include your Python code and output in the code box below.)


import tensorflow as tf
import matplotlib.pyplot as plt
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten
from tensorflow.keras.datasets import fashion_mnist

# Load dataset
(x_train, y_train), (x_test, y_test) = fashion_mnist.load_data()

# Normalize data
x_train = x_train / 255.0
x_test = x_test / 255.0

# Build model
model = Sequential([
    Flatten(input_shape=(28, 28)),
    Dense(128, activation='relu'),
    Dense(10, activation='softmax')
])

# Compile model
model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

# Train model
history = model.fit(
    x_train, y_train,
    epochs=10,
    validation_data=(x_test, y_test)
)

# Plot Accuracy
plt.figure()
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title("Accuracy Curve")
plt.xlabel("Epochs")
plt.ylabel("Accuracy")
plt.legend(["Training", "Validation"])
plt.show()

# Plot Loss
plt.figure()
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title("Loss Curve")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.legend(["Training", "Validation"])
plt.show()
```
